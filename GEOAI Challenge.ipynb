```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import geopandas as gpd
import rasterio
from rasterstats import zonal_stats
import os
from shapely.geometry import Point


# Set random seed for reproducibility
SEED = 42
np.random.seed(SEED)
```

```python
# Full base path
DATA_PATH = r"C:\Users\janen\Desktop\GEOAI Challenge for Cropland Mapping in Dry Environments\data"

# Load the shapefiles
fergana_gdf = gpd.read_file(os.path.join(DATA_PATH, "train", "Fergana_training_samples.shp"))
orenburg_gdf = gpd.read_file(os.path.join(DATA_PATH, "train", "Orenburg_training_samples.shp"))

# Assign CRS manually if missing
fergana_gdf = fergana_gdf.set_crs("EPSG:4326")
orenburg_gdf = orenburg_gdf.set_crs("EPSG:4326")

# Combine both training regions using pd.concat
gdf_train = pd.concat([fergana_gdf, orenburg_gdf], ignore_index=True)
```

```python
s1 = pd.read_csv(os.path.join(DATA_PATH, 'Sentinel1.csv'), nrows=100000)
s2 = pd.read_csv(os.path.join(DATA_PATH, 'Sentinel2.csv'), nrows=100000)
```

```python
s1["geometry"] = s1.apply(lambda row: Point(row["translated_lon"], row["translated_lat"]), axis=1)
gdf_sentinel1 = gpd.GeoDataFrame(s1, geometry="geometry", crs="EPSG:4326")
gdf_sentinel1 = gdf_sentinel1.to_crs(gdf_train.crs)
```
